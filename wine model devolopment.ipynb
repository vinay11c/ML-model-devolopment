```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x=wine_df.drop(['class_0'],axis=1)
y=wine_df['class_0']
x.columns
```

```python
vif_scale=[]
for i in range(len(x.columns)):
    vif_scale.append(variance_inflation_factor(scaled,i))
vif_scale
```

```python
vif_scaled_df=pd.DataFrame(vif_scale,index=x.columns,columns=['VIF'])
vif_scaled_df.sort_values(by='VIF',ascending=False)
```

```python
col1=vif_scaled_df[vif_scaled_df['VIF']<15].reset_index()['level_0'].tolist()
col1
```

```python
col2=vif_scaled_df[vif_scaled_df['VIF']>15].reset_index()['level_0'].tolist()
col2
```

```python
finalcols=col1+col2
```

```python
final_df=wine_df[finalcols]
final_df['class_0']=y
final_df
```
